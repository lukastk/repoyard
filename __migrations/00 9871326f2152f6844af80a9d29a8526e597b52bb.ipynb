```python
repo_paths = sorted(
    list((Path.home() / ".repoyard" / "local_store" / "hetzner-box").iterdir())
)

dev_path = Path.home() / "dev"
dev_path.mkdir(parents=True, exist_ok=True)

for repo_path in repo_paths:
    if repo_path.is_file():
        continue

    dest_path = dev_path / repo_path.name

    if dest_path.exists():
        raise FileExistsError(f"Destination path {dest_path} already exists")

    shutil.move(repo_path / "data", dest_path)
```
